```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('taxis')
```

```python
df.iloc[[0,4],[2,3,4]]
```

```python
df.loc[0:20,['passengers','fare']]
df.set_index('pickup',inplace=True)
```

```python
df.loc['2019-03-04 16:11:55',['passengers','fare']]
```

```python
df.index
```

```python
df.loc[['2019-03-23 20:21:09', '2019-03-04 16:11:55',
               '2019-03-27 17:53:01', '2019-03-10 01:23:59',
               '2019-03-30 13:27:42', '2019-03-11 10:37:23',
               '2019-03-26 21:07:31', '2019-03-22 12:47:13',
               '2019-03-23 11:48:50', '2019-03-08 16:18:37'],['fare','total']]
```
